```python
from bs4 import BeautifulSoup as bs
import requests

url = 'https://www.yelp.com/search?find_desc=&find_loc=Saint%20Louis%2C%20MO%2063104'
```

```python
response = requests.get(url)
response
```

```python
soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
print(soup.title)
```

Gather links from first page

```python
all_links = []
for item in soup.select('h4'): 
    try:  
        if item.find('a'):
            href = item.find('a', href = True)
            all_links.append(href['href'])
    except Exception as e: 
        raise e 
        print('')
```

```python
all_links
```

Gather links for first fifty restaurants

```python
all_urls = []
for i in range(0,50,10):
    url = f'https://www.yelp.com/search?find_desc=&find_loc=Saint%20Louis%2C%20MO%2063104&start={i}'
    response = requests.get(url)
    soup = bs(response.text, 'html.parser')
    for item in soup.select('h4'): 
        try:  
            if item.find('a'):
                href = item.find('a', href = True)
                all_urls.append(href['href'])
        except Exception as e: 
            raise e 
            print('')
```

### What is web scraping? Why is it helpful? Why is it sometimes in a legal grey area?

Web scraping is using a program to automatically pull information or data from websites, or to automatically fill in forms on websites.  Web scraping is helpful because it is often used to automate tasks which are time intensive or tedious.  It can allow for things to be accomplished much faster.  Web scraping is a legal grey area because some courts have upheld that if data is avialable to the public it is fair use to scrape, while other courts have ruled against scraping by defining what is considered public data in more restrictive ways.
